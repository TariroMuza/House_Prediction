### Task 2 GridSearchCV_ Regression

```python
models=[]   # model[] is empty
scores=[]   #scores[] is empty
```

```python
# Import relavent libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV

# Define the parameters for grid search
parameters = {'fit_intercept': [True, False]}

# Initialize the LinearRegression model
c7 = LinearRegression()

# Initialize GridSearchCV with the model and parameters
grid = GridSearchCV(c7, parameters, cv=3)

# Fit the model to the data
grid.fit(x, y)

# Print the best parameters and the best estimator
print("Best Parameters:", grid.best_params_)
print("Best Estimator:", grid.best_estimator_)
print("Best Score:", grid.best_score_)

# Append the best estimator to the models list
models.append(grid.best_estimator_)

# Append the best score to the scores list
scores.append(grid.best_score_)
```

```python
# Create a dataframe to store results
data=pd.DataFrame({"model_name":models,"best_score":scores})
```

```python
print(data)
```

The results of the GridSearchCV indicate the following:

Best Parameters= ('fit_intercept': True),
Best Estimator= LinearRegression(),
Best Score= 0.7031337342269555
This suggests that using LinearRegression with fit_intercept=True resulted in the highest cross-validated score, which is approximately 0.7031

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# Define the parameter grid with the values you want to test
param_grid = {'n_neighbors': [3, 5, 7, 9, 11],
                'metric': ['manhattan', 'euclidean']}

# Initialize the KNeighborsClassifier
knn = KNeighborsClassifier()

# Initialize GridSearchCV with the classifier, parameter grid, and cross-validation strategy
grid_search = GridSearchCV(knn, param_grid, cv=5)

# Fit the grid search to  training data
grid_search.fit(x_train, y_train)

# Get the best estimator found by the grid search
best_knn = grid_search.best_estimator_

# Use the best estimator to make predictions
y_pred_KNN = best_knn.predict(x_test)

# Print the classification report
print(classification_report(y_test, y_pred_KNN))

# Optionally, print the best parameters
print(f"Best parameters found: {grid_search.best_params_}")
```

These results suggest that the model performs well with high accuracy, precision, recall, and F1-scores, especially for class 0, while its performance on class 1 is still strong but slightly less effective in terms of recall. The best parameters indicate that a KNN model with 7 neighbors and Manhattan metric was found to be optimal for this classification task.
